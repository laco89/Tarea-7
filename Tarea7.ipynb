### Primer metodo

#### Define arrays para condiciones iniciales y define el $\Delta t$ y $\Delta x$

```python
final=py.zeros(200)
c=1
delta_t=0.01
delta_x=0.01005025
beta=c*delta_t/delta_x
```

#### Define la funcion del esquema1

```python
def esquema1 (pinicial):
    lista = py.zeros(200)
    lista[0]=pinicial[0];
    for i in range(1,199):
        lista[i]=pinicial[i]-beta*(pinicial[i+1]**2-pinicial[i-1]**2)/2
    lista[199]=pinicial[199]
    return lista
```

#### Toma condicion inicial con velocidad del sonido c=1 y la grafica

```python
x = linspace(0, 2.0, 200)
inicial=3*sin(pi*x)
inicial[199]=0
plt.plot(x,inicial)
```

#### Corre el esquema1 en pasos de $\Delta t$ hasta que el tiempo total sea 0.15

```python
z=py.array([inicial])
T=0

while T<0.15:
    final=esquema1(inicial)
    z=py.append(z,[final],0)
    
    inicial=final
    
    T=T+delta_t
size(z)
```

# Segundo metodo, esquema Lax-Wendroff

#### Se inician los arrays con 200 items para las condiciones iniciales y se define el esquema Lax-Wendroff

```python
def esquema2 (pinicial):
    lista = py.zeros(200)
    lista[0]=pinicial[0];
    for i in range(1,199):
        lista[i]=pinicial[i]-(beta/4)*((pinicial[i+1])**2-(pinicial[i-1])**2)+(beta**2)/8*((pinicial[i+1]+pinicial[i])*(pinicial[i+1]**2-pinicial[i]**2)-(pinicial[i]+pinicial[i-1])*(pinicial[i]**2-pinicial[i-1]**2))
        
        
        
    lista[199]=pinicial[199]
    return lista
```

#### Define y grafica de las condiciones iniciales

```python
x = linspace(0, 2.0, 200)
inicial=3*sin(pi*x)
inicial[199]=0
plt.plot(x,inicial)
```

#### Se usa el mismo $\Delta t$ usado en el metodo anterior para correr el segundo esquema en pasos de $\Delta t$

```python
z=py.array([inicial])
T=0

while T<0.15:
    final=esquema2(inicial)
    z=py.append(z,[final],0)
    
    inicial=final
    
    T=T+delta_t
size(z)
```

Se puede observar que el metodo de Lax-Wendroff es mas exacto, por lo tanto este es mejor para trabajarlo.
